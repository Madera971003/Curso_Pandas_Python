```python
df = pd.DataFrame(np.arange(0, 15).reshape(5, 3), columns=['a', 'b', 'c'])
df
```

```python
df['d'] = np.nan
df['e'] = np.arange(15, 20)
df.loc[5,:] = pd.NA
df.loc[4,'a'] = pd.NA
df.loc[0,'d'] = 1
df.loc[5,'d'] = 10
df
```

```python
df.isnull() #Para reconocer cuando un objeto es nulo
```

```python
df.isna() #Cumple con la funcion anterior
```

```python
df.isnull().sum() #Encuentra y suma la cantidad de objetos nulos
```

```python
df.notnull().sum(axis=1) #conocer el número de filas con elementos nulos, se usa el axis = 1
```

```python
df.size-df.isnull().sum().sum() #La suma de todos los elementos nulos en el DF
```

```python
df[df['a'].notnull()] #filtrar por las variables no nulas de la columna ‘a’
```

```python
df.dropna() #elimna rapidamente las filas con registros faltantes
```

```python
df[['a']].dropna()
```

```python
df.fillna(0) #Rellenar los valores nulos con algo; en este caso 0
```

```python
df.fillna(method="ffill") #Se rellena con el valor siguiente; forma descendente; en direccion de columnas
```

```python
df.fillna(method="bfill") #Se rellena con el valor previo; forma ascendente en direccion de columnas
#Solo se rellena con valores de 10; ya que no hay mas valores en una 6 fila
```

```python
df.fillna(method="bfill",axis=1) #Lo mismo a lo anterior pero en direccion de las filas
```

```python
fill = pd.Series([100, 101, 102]) #Podemos usar también una serie para reemplazar los valores de una columna en especifico
# es importante que haya emparejamiento entre los índices
fill
```

```python
df['d'] = df['d'].fillna(fill) #Intersa ek valor de fill si el del df es nulo
df['d'] 
```

```python
df
```

```python
df.fillna(df.median()) #Rellena los valores nulos con los promedios (tambien se puede usar mean())
```

Pandas también puede interporlar los valores faltanes calculando el valor que puede haber existido en el medio

```python
df_d = pd.concat([df[['d']], df[['d']].interpolate()],axis=1)
df_d.columns = ['d_antes','d_interpolado']
df_d
```
