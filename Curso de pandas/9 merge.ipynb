```python
import pandas as pd
pd.__version__
```

```python
df_left = pd.DataFrame(
    {'X':['x0','x1','x2','x3'],
    'W':['w0','w1','w2','w3'],
    'Y':['y0','y1','y2','y3'],
    'Mix':['y2','y3','a2','a3']},
    index = ['y2','y3','a2','a3'])
df_left
```

```python
df_right = pd.DataFrame(
    {'Z':['z2','z3','z4','z5'],
     'A':['a2','a3','a4','a5'],
     'Y':['y2','y3','y4','y5']},
    index = [2,3,4,5])
df_right
```

```python
pd.merge(df_left, df_right) #Merge entre DF con los parametros compartidos, en este caso y2 y y3
```

```python
pd.merge(df_left, df_right, how = 'inner', on = 'Y') #Paso lo mismo a lo anterior
```

```python
pd.merge(df_left, df_right, how = 'inner', left_on = 'Mix', right_on = 'Y') #Se unen a traves de los parametros Y
```

```python
pd.merge(df_left, df_right, how = 'inner', left_on = 'Mix', right_on = 'A') #Se unen a traves de los parametros A
```

```python
df_left = pd.DataFrame(
    {'X':['x0','x1','x2','x3'],
    'W':['w0','w1','w2','w3'],
    'Y':['y0','y1','y2','y3']},
    index = ['y2','y3','a2','a3'])
df_left
```

```python
pd.merge(df_left, df_right, how = 'inner', on = 'Y')
```

```python
pd.merge(df_left, df_right, how = 'left', on = 'Y')
```

```python
pd.merge(df_left, df_right, how = 'right', on = 'Y')
```

```python
pd.merge(df_left, df_right, how = 'outer', on = 'Y')
```

```python
df_left = pd.DataFrame(
    {'X':['x0','x1','x2','x3'],
    'W':['w0','w1','w2','w3'],
    'Y':['y0','y1','y2','y3'],
    'A':['a0','a1','a2','a3']}
    )
df_left
```

```python
df_right
```

```python
pd.merge(df_left, df_right, how = 'outer', on = ['Y', 'A']) #Aplica merge a traves de A y Y
```

```python
pd.merge(df_left, df_right, how = 'outer', on = 'A') #Aqui se obtienen mas datos que la anterior
```

Los sufijos se agregan automticamente; pero estos pueden ser cambiados

```python
pd.merge(df_left, df_right, how = 'outer', on = 'A', suffixes=['_left', '_right']) 
```
