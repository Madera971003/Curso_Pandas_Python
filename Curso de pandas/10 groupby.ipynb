```python
def mean_eur2usd(x):
  return np.mean(x)*1.12
```

```python
mean_eur2usd([100, 102]) #Prueba de la funcion
```

```python
df.groupby('sex')[['total_bill', 'prct_tip']].apply(mean_eur2usd) #Aplicacion de la funcion creada
```

```python
df.groupby(['sex', 'time'])[['total_bill', 'prct_tip']].apply(mean_eur2usd) #Aqui se aplica una funcion a una doble agrupacion
```

```python
df.groupby(['sex', 'time'])[['total_bill', 'prct_tip']].apply(np.std) #Variacion estandar
```

```python
df.groupby(['sex', 'time'])[['total_bill', 'prct_tip']].apply(lambda x: np.mean(x)*1.12)
#Realiza lo mismo que mean_eur2usd() pero con lamdas
#x: este es una lista
```

```python
#Para aplicar mas de una funcion se usa aggregate o su version corta: agg
df.groupby(['sex', 'time'])[['total_bill', 'prct_tip']].aggregate([np.mean, np.std, np.max, np.min, lambda x: np.mean(x)*1.12, mean_eur2usd])
```

```python
#Agregar funciones a traves de diccionarios; este contendrÃ¡ un conjunto de funciones
dict_agg = {'tip': [min, max], 'total_bill': [np.mean, mean_eur2usd, lambda x: np.mean(x)*1.12], 'prct_tip': [np.std]}
dict_agg
```

```python
df.groupby(['sex', 'time'])[['tip', 'prct_tip', 'total_bill']].agg(dict_agg)
#En este caso importarÃ¡ el orden entre tip y prct_tip; ya que uno depende de otro; el resto no tendra relevancia el orden
```

```python
def f_filter(x):
  return mean_eur2usd(x['total_bill'].mean()) > 20
  #Este calcula el promedio de la columna agrupada de total_bill
  #Luego hace la conversion con la funcion mean_eur2usd
  #Por ultimo recoge los datos mayores a 20
```

```python
df_filtered = df.groupby(['sex', 'time']).filter(f_filter)
df_filtered
```

```python
df_filtered.groupby(['sex', 'time']).count()
```

```python
df['ones'] = 1
df
```

```python
df_g = df.groupby(['sex', 'smoker'])[['ones']].sum() #Se agrupan y se suman la cantidad de cada grupo
df_g
```

```python
df_g.groupby(level = 0).apply( #Agrupacion por uno de sus indices; en este caso en el nivel del genero
    lambda x:
      x/x.sum() * 100 #Solucion ej: 60/(60+97)
)
```

```python
pd.cut(df['total_bill'], bins = 3) #Categorizado en intervalos; bins: es el numero de categorias
```

```python
df.groupby(['time', 'bin_total'])[['ones']].count()
```

```python
df.groupby(['time', 'bin_total'])[['ones']].count().groupby(level = 0).apply(
    lambda x:
      x/x.sum() * 100
)
```

**Pivot**: Tabla dinÃ¡mica (es de pandas)

```python
df.groupby(['sex', 'time'])[['total_bill']].mean()
```

```python
df_gp = df.groupby(['sex', 'time'])[['total_bill']].mean().reset_index()
df_gp
```

```python
df_gp.pivot_table(values = 'total_bill', index = 'sex', columns = 'time') #Se modifica la estructura de la tabla
#De la anterior a esta
```

![](https://pandas-docs.github.io/pandas-docs-travis/_images/reshaping_pivot.png)

```python
df.pivot_table(values = 'total_bill', index = 'sex', columns = 'time') #Internamente Pivot table hace el promedio de los datos
```

```python
df.pivot_table(values = 'total_bill', index = 'sex', columns = 'time', aggfunc=np.median)
#Con aggfuc puedo cambiar la funcion por default
```

```python
df_pivot = df.pivot_table(values = 'total_bill', index = 'sex', columns = 'time', aggfunc=[np.median, np.std])
df_pivot
```

```python
df_pivot.unstack() #Se eliminan las categorÃ­as
```

```python
df_pivot.unstack().reset_index() #Resetea los indices
```
