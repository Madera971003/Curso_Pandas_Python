```python
import pandas as pd
import numpy as np
pd.__version__
```

```python
df_lmerged = pd.read_csv('london_merged.csv')
df_lmerged
```

```python
df_lmerged.dtypes
```

```python
df_lmerged['timestamp'] = pd.to_datetime(df_lmerged['timestamp']) #Cambiar formato de fechas
df_lmerged
```

```python
df_lmerged['hour'] = df_lmerged['timestamp'].dt.hour
df_lmerged['hour']
```

```python
df = df_lmerged.iloc[:, 1:]
df
```

```python
df['wind_speed']**2 #Eleva al cuadrado los datos de wind_speed
```

```python
df['wind_speed']**2 +10
```

```python
np.sin(df['wind_speed']**2) + 10
```

```python
df['t1'] - df['t2']
```

```python
df['t1'].iloc[::2] - df['t2'] #Selecciona los datos pares de 2 en 2
```

```python
df['t1'].iloc[::2].sub(df['t2'], fill_value = 1000) #Rellena los Nan con un 1000 y aplica resta
```

```python
df['t1'] / df['t2']
```

```python
df['t1'].dot(df['t2']) #Producto punto
```

```python
def fun_1(x): #Declaracion de funcion
  y = x**2 + 1
  return y
```

```python
fun_1(10)
```

```python
np.arange(-5, 6)
```

```python
fun_1(np.arange(-5, 6))
```

```python
df['hour'].apply((fun_1)) #Aplica cierta funcion a determinada columna
```

```python
def fun_2(x, a=1, b=0): #Declaracion de funcion
  y = x**2 + a*x + b
  return y
```

```python
fun_2(10)
```

```python
fun_2(10, a = 20, b = -4)
```

```python
df['hour'].apply(fun_2, args = (20, -100)) #Se aplica la funcion 2 agregando los demas argumentos
```

```python
df['hour'].apply(fun_2, a = 20, b = -100) #Otra forma de incluir los argumentos
```

```python
df['t1'].apply(lambda x: x+273) #Uso de High Order Functions lambdas
```

```python
df.apply(lambda x: x.mean()) #Halla el promedio de cada columna
```

```python
df.apply(lambda x: x.mean(), axis = 1) #Se calcula el promedio pero ahora en el eje de las filas
```

```python
df.apply(lambda x: x.std(), axis = 1) #Calculo de la desviacion estandar en filas
```

```python
df.apply(lambda x: x['t1']-x['t2'], axis = 1) #Diferencia de temperaturas
```

```python
df.applymap(lambda x: x/1000) #Se aplica la funcion a todas las celdas de la tabla
```

```python
df['wind_speed'].map('The wind speed is {}'.format) #Usar map para imprimir un mensaje de la siguiente manera
```
