```python
import pandas as pd
```

```python
df = pd.DataFrame({'a': ['w'] * 4 + ['x'] * 3 + ['y'] * 2 + ['z']+['v'], 
                   'b': [1, 1, 1, 1, 2, 2, 2, 3, 3, 4,5]})
df
```

```python
df.duplicated() #Sirve para encontrar registros duplicados, sin tomar en cuenta uno de ellos
```

```python
df.duplicated(keep='first') #usar keep='first' para marcar solo la primera ocurrencia o keep='last' para marcar la Ãºltima
```

```python
df.duplicated(keep='last')
```

```python
df[~ df.duplicated()] #Filtro para obtener los datos no duplicados ~: es la negacion, pera evitar ver los duplicados
#Se puede usar keep con first o last, dependiendo el caso
```

```python
df.duplicated(keep=False) #Este muestra cuales registros estan duplicados tomando el original
```

```python
df[df.duplicated(keep=False)] #Muestra los datos de quienes son los duplicados, omitiendo los que no
```

```python
df.drop_duplicates() #Elimina los registros duplicados; por defecto keep='first'
```

```python
df.drop_duplicates(['a'],keep='last') #Eliminar duplicados tomando en cuenta los duplicados en una columna
```
