```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('covid_19_data.csv')
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])
df.sample(10)
```

```python
df = df[['ObservationDate',
 'Country/Region',
 'Confirmed',
 'Deaths',
 'Recovered']]
df
```

```python
df_time  = df.groupby(['Country/Region',
           pd.Grouper(key = 'ObservationDate', freq = '1D')]).sum()
df_time
```
