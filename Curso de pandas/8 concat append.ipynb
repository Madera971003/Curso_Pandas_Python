```python
x1 = np.random.rand(2,5) * 10 #Generacion de datos
x1
```

```python
x2 = np.random.rand(2,5) * -1
x2
```

```python
np.concatenate([x1,x2]) #Concatena ambas matrices, creando m√°s filas; numpy
```

```python
np.concatenate([x1,x2]).shape
```

```python
np.concatenate([x1,x2], axis = 1) #Se concatenan creandose mas columnas; numpy
```

```python
np.concatenate([x1,x2], axis = 1).shape
```

```python
s1 = pd.Series(x1[0], index = ['a', 'b', 'c', 'd', 'e']) #Serie de datos con pandas
s1
```

```python
s2 = pd.Series(x2[0], index = ['c', 'd', 'e', 'f', 'g'])
s2
```

```python
pd.concat([s1,s2]) #Se concatenan; pandas
```

```python
pd.concat([s1,s2], axis = 1) #Se concatenan respetando los indices
```

```python
#Si no se requiere concatenar con base a los indices, conviene resetearlos
s1.reset_index(drop=True) #Sin la funcion drop quedar√° existiendo el indice como una columnas
```

```python
pd.concat([s1.reset_index(drop=True),s2.reset_index(drop=True)], axis = 1)
```

```python
df1 = pd.DataFrame(np.random.rand(3,2)*10, columns=['a', 'b'])
df1
```

```python
df2 = pd.DataFrame(np.random.rand(3,2)*-1, columns=['a', 'b'], index = [2, 3, 4])
df2
```

```python
pd.concat([df1, df2]) #Concatenar DataFrame, se repite un indice
```

```python
pd.concat([df1, df2], axis = 1) #Se concatenan a lo largo de las columnas
```

```python
pd.concat([df1, df2], axis = 1, join='inner') #Se concatenan los datos con indices compartidos
```

```python
pd.concat([
           df1.reset_index(drop=True),
           df2.reset_index(drop=True)
           ], axis = 1) #Se resetean los indices para la concatenacion
```

```python
df1.append(df2) #Otra forma de concatenar df
```

```python
df1.append(df2).append(df2) #Concatenacion tras concatenacion
```

```python
df1.T #Funcion de transpuesta
```

```python
df1.T.append(df1.T).T
```
